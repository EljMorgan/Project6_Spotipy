```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import xml
```

```python
url='https://en.wikipedia.org/wiki/List_of_music_genres_and_styles'
```

```python
page= requests.get(url)
soup=bs(page.content,'lxml')
html=soup.find_all('div', class_='div-col')
genres=[]
for i in range(len(html)):
    html[i].find_all('li')
    for g in range(len(html[i].find_all('li'))):
        genres.append(html[i].find_all('li')[g].text)
#genres=[l.text for l in html]
    
#genres[0].split('\n','n')

print(len(genres))
genres
genres2=[]
for g in genres:
    if '\n' in g:
        g.split(sep='\n')
        for g in g.split(sep='\n'):
            genres2.append(g)
        
    else:
        genres2.append(g)
    
print(len(genres2))
print(genres2)

```
